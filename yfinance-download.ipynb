# Fetches sector weightings for a list of ETFs 

```python
import yfinance as yf
import pandas as pd
import numpy as np

def get_sector_weights(etfs):
    """
    Fetches sector weightings for a list of ETFs using yfinance and returns a pandas DataFrame
    with sectors as rows and ETFs as columns. Missing sectors are filled with 0.
    
    Parameters:
    etfs (list): List of ETF ticker symbols.
    
    Returns:
    pd.DataFrame: DataFrame with sector weights.
    """
    sector_data = {}
    for etf in etfs:
        ticker = yf.Ticker(etf)
        funds_data = ticker.funds_data
        sector_weightings = funds_data.sector_weightings
        sector_data[etf] = sector_weightings
    
    df = pd.DataFrame(sector_data).fillna(0)
    return df

def get_top_holdings(etfs):
    """
    Fetches sector weightings for a list of ETFs using yfinance and returns a pandas DataFrame
    with sectors as rows and ETFs as columns. Missing sectors are filled with 0.
    
    Parameters:
    etfs (list): List of ETF ticker symbols.
    
    Returns:
    pd.DataFrame: DataFrame with sector weights.
    """
    top_holdings_data = {}
    for etf in etfs:
        ticker = yf.Ticker(etf)
        funds_data = ticker.funds_data
        top_holdings = funds_data.top_holdings
        top_holdings_data[etf] = top_holdings['Holding Percent']
    
    df = pd.DataFrame(top_holdings_data).fillna(0)
    return df

def calculate_portfolio_sector_weights(etfs, weights):
    """
    Calculates the weighted sector allocations for a portfolio of ETFs.
    
    Parameters:
    etfs (list): List of ETF ticker symbols.
    weights (list): List of weights (as decimals) corresponding to each ETF.
    
    Returns:
    pd.Series: Weighted sector allocations for the portfolio.
    """
    # Get sector weights DataFrame
    sector_df = get_sector_weights(etfs)
    
    # Convert weights to a numpy array for calculation
    weights_array = np.array(weights)
    
    # Calculate weighted sector allocations
    weighted_sectors = sector_df.dot(weights_array)
    
    return weighted_sectors

# ETF tickers and their corresponding weights
etfs = ['VPU', 'VXF', 'VO', 'VGT', 'VHT', 'VCR', 'VBK', 'VAW', 'VB', 'VUG', 'VIS', 'VBR', 'VONG', 'EZBC']
weights = [0.163, 0.153, 0.148, 0.116, 0.112, 0.084, 0.069, 0.046, 0.035, 0.03, 0.019, 0.015, 0.009, 0.001]

# Calculate portfolio sector weights
portfolio_sectors = calculate_portfolio_sector_weights(etfs, weights)
top_holdings_df = get_top_holdings(etfs)
# Print results
print("Portfolio Sector Weights:")
print(portfolio_sectors.sort_values(ascending=False))
```
